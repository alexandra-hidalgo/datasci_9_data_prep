```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
NYPD = pd.read_csv('NYPD_shooting_data.csv')
NYPD
```

```python
NYPD.columns
```

```python
NYPD.columns = NYPD.columns.str.lower()
NYPD.columns
```

```python
NYPD.shape
```

```python
NYPD.dtypes
```

```python
NYPD.isnull().sum()
```

```python
# Removing columns with missing values
to_drop = [
    'loc_of_occur_desc',
    'jurisdiction_code',
    'loc_classfctn_desc',
    'location_desc',
    'perp_age_group',
    'perp_sex',
    'perp_race',
    'latitute',
    'longitude',
    'lon_lat',
]
```

```python
NYPD.drop(to_drop, axis=1, inplace=True, errors='ignore')
NYPD.shape
```

## Data transformation.

```python
# Date
NYPD['occur_date'] = pd.to_datetime(NYPD['occur_date']).dt.date
```

```python
# Now encode occur_date so it is a day of the week
NYPD['occur_date'] = pd.to_datetime(NYPD['occur_date'])
NYPD['occur_date'] = NYPD['occur_date'].dt.day_name()
NYPD.occur_date
```

```python
# performing encoding on date
enc = OrdinalEncoder()
enc.fit(NYPD[['occur_date']])
NYPD['occur_date']= enc.transform(NYPD[['occur_date']])
NYPD.occur_date
```

```python
## create dataframe with mapping
NYPD_mapping_date = pd.DataFrame(enc.categories_[0], columns=['occur_date'])
NYPD_mapping_date['occur_date_ordinal'] = NYPD_mapping_date.index
NYPD_mapping_date
```

```python
## boro --> will need to encode this
NYPD.boro.value_counts()
```

```python
## perform orindla encoding on boro
enc = OrdinalEncoder()
enc.fit(NYPD[['boro']])
NYPD['boro'] = enc.transform(NYPD[['boro']])
NYPD.boro
```

```python
## create dataframe with mapping
NYPD_mapping_area = pd.DataFrame(enc.categories_[0], columns=['boro'])
NYPD_mapping_area['boro_ordinal'] = NYPD_mapping_area.index
NYPD_mapping_area
```

```python
## statistical murder flag --> will need to encode this
NYPD.statistical_murder_flag.value_counts()
```

```python
## performing ordinal encoding on statistical murder flag
enc = OrdinalEncoder()
enc.fit(NYPD[['statistical_murder_flag']])
NYPD['statistical_murder_flag'] = enc.transform(NYPD[['statistical_murder_flag']])
NYPD.statistical_murder_flag
```

```python
## create dataframe with mapping
NYPD_mapping_flag = pd.DataFrame(enc.categories_[0], columns=['statistical_murder_flag'])
NYPD_mapping_flag['SMF_ordinal'] = NYPD_mapping_flag.index
NYPD_mapping_flag
```

```python
## victim age group needs to be encode
NYPD.vic_age_group.value_counts()
```

```python
## perform ordinal encoding on vic_age_group
enc = OrdinalEncoder()
enc.fit(NYPD[['vic_age_group']])
NYPD['vic_age_group'] = enc.transform(NYPD[['vic_age_group']])
NYPD.vic_age_group
```

```python
## create dataframe with mapping
NYPD_mapping_age_group = pd.DataFrame(enc.categories_[0], columns=['vic_age_group'])
NYPD_mapping_age_group['vic_age_group'] = NYPD_mapping_age_group.index
NYPD_mapping_age_group
```

```python
# Victim sex needs to be encode
NYPD.vic_sex.value_counts()
```

```python
## drop row if sex is equal to U
NYPD = NYPD[NYPD['vic_sex'] != 'U' ]
NYPD.vic_sex.value_counts()
```

```python
## perform ordinal encoding on vic_sex
enc = OrdinalEncoder()
enc.fit(NYPD[['vic_sex']])
NYPD['vic_sex'] = enc.transform(NYPD[['vic_sex']])
NYPD.vic_sex.value_counts()
```

```python
## create dataframe with mapping
NYPD_mapping_sex = pd.DataFrame(enc.categories_[0], columns=['vic_sex'])
NYPD_mapping_sex['vic_sex_ordinal'] = NYPD_mapping_sex.index
NYPD_mapping_sex
```

```python
## vic_race
NYPD.vic_race.value_counts()
```

```python
## create dataframe with mapping
NYPD_mapping_vic_race = pd.DataFrame(enc.categories_[0], columns=['vic_race'])
NYPD_mapping_vic_race['vic_race_ordinal'] = NYPD_mapping_vic_race.index
NYPD_mapping_vic_race
descent_mapping = {
    'A': 'ASIAN / PACIFIC ISLANDER',
    'B': 'BLACK',
    'G': 'BLACK HISPANIC',
    'H': 'WHITE HISPANIC',
    'I': 'AMERICAN INDIAN/ALASKAN NATIVE',
    'U': 'UNKOWN',
    'W': 'WHITE'
}
NYPD_mapping_vic_race['vic_race_desc'] = NYPD_mapping_vic_race['vic_race'].map(descent_mapping)
```
